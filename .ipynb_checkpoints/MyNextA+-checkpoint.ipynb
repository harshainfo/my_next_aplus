```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Oct 10 06:27:05 2019

@author: hpitawela
"""

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
import keras

#importing dataset
dataset = pd.read_csv("training.psv",delimiter='|')
dataset_cropped = dataset[['course','level']].copy()
dataset_cropped2 = dataset_cropped.copy()
#dataset_cropped2.set_index(['course','level'], inplace=True)
#index = pd.Index(dataset_cropped2.index)
dataset_cropped2.drop_duplicates(keep='first', inplace=True)
dataset_cropped2.sort_values("course", inplace=True)
#df2 = pd.DataFrame(index)
df_grouped = dataset_cropped2.groupby(['course']).count()

courses_senior = dataset_cropped2.loc[dataset_cropped2.level=='Senior',:]
courses_senior.reset_index(drop=True, inplace=True)
courses_senior.set_index(['course'], inplace=True)
courses_junior = dataset_cropped2.loc[dataset_cropped2.level=='Junior',:]
courses_junior.reset_index(drop=True, inplace=True)
courses_junior.set_index(['course'], inplace=True)
courses_sopho = dataset_cropped2.loc[dataset_cropped2.level=='Sophomore',:]
courses_sopho.reset_index(drop=True, inplace=True)
courses_sopho.set_index(['course'], inplace=True)
courses_fresh = dataset_cropped2.loc[dataset_cropped2.level=='Freshman',:]
courses_fresh.reset_index(drop=True, inplace=True)
courses_fresh.set_index(['course'], inplace=True)

pure_senior_index = courses_senior.index.difference(courses_junior.index).difference(courses_sopho.index).difference(courses_fresh.index)
print(pure_senior_index.shape)
pure_junior_index = courses_junior.index.difference(courses_senior.index).difference(courses_sopho.index).difference(courses_fresh.index)
print(pure_junior_index.shape)
pure_sopho_index = courses_sopho.index.difference(courses_senior.index).difference(courses_junior.index).difference(courses_fresh.index)
print(pure_sopho_index.shape)
pure_fresh_index = courses_fresh.index.difference(courses_senior.index).difference(courses_junior.index).difference(courses_sopho.index)
print(pure_fresh_index.shape)
print(pure_senior_index)

input_index = pure_fresh_index.union(pure_sopho_index).union(pure_junior_index)
input_index_values = input_index.values
#print(input_index_values)

output_index_values = pure_senior_index.values

fname_train = 'training.psv'
fname_eval = 'eval.psv'

df_train = pd.read_csv(fname_train, delimiter="|")
df_eval = pd.read_csv(fname_eval, delimiter="|")

# grouping by student_id to extract majors
df2_train = df_train.groupby('student_id', as_index=False).last()
df2_train = df2_train[['student_id','major']]

# one hot encoding of majors in the training set
encoded_majors = pd.get_dummies(df2_train.major)
df2_train_st_major_headers = pd.concat([df2_train, encoded_majors], axis=1)
df2_train_major_headers = df2_train_st_major_headers.drop(columns=['major'])

# set student_id as index
df2_train_majors_with_st_index = df2_train_major_headers.set_index('student_id')
df2_train_majors_with_st_index.head()


# grade values in training set
df1_train_grade = df_train.pivot(index='student_id',columns='course',values='grade')

print(df1_train_grade.head())

df_input = df1_train_grade.loc[:,input_index_values]

print(df_input.head())

df_output = df1_train_grade.loc[:,output_index_values]

print(df_output.head())

print(pd.unique(df_input[df_input.columns].values.ravel('K')))
print(pd.unique(df_output[df_output.columns].values.ravel('K')))

# One hot encoding for grades in training set
df_input_grade_course_headers = pd.concat([pd.get_dummies(df_input[col]) for col in df_input], axis=1, keys=df_input.columns)
df_output_grade_course_headers = pd.concat([pd.get_dummies(df_output[col]) for col in df_output], axis=1, keys=df_output.columns)

print(df_input_grade_course_headers.head())
print(df_output_grade_course_headers.head())

print(pd.unique(df_input_grade_course_headers[df_input_grade_course_headers.columns].values.ravel('K')))
print(pd.unique(df_output_grade_course_headers[df_output_grade_course_headers.columns].values.ravel('K')))

from sklearn.impute import SimpleImputer
in_imputer = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=0, verbose=0, copy=False)
out_imputer = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=0, verbose=0, copy=False)
input_imputer = in_imputer.fit(df_input_grade_course_headers.values[:,1:394])
output_imputer = out_imputer.fit(df_output_grade_course_headers.values[:,1:959])

df_input_grade_course_headers.values[:,1:394] = in_imputer.transform(df_input_grade_course_headers.values[:,1:394])
df_output_grade_course_headers.values[:,1:959] = out_imputer.transform(df_output_grade_course_headers.values[:,1:959])

print(pd.unique(df_input_grade_course_headers[df_input_grade_course_headers.columns].values.ravel('K')))
print(pd.unique(df_output_grade_course_headers[df_output_grade_course_headers.columns].values.ravel('K')))

df_majors = dataset[['student_id','major']].copy()

encoded_majors = pd.get_dummies(df_majors.major)
df2_train_st_major_headers = pd.concat([df_majors, encoded_majors], axis=1)
df2_train_major_headers = df2_train_st_major_headers.drop(columns=['major'])

# set student_id as index
df2_train_majors_with_st_index = df2_train_major_headers.set_index('student_id')
df2_train_majors_with_st_index.head()



```

```python
result = pd.concat([df_input_grade_course_headers, df2_train_majors_with_st_index], axis=1, sort=False)
```

```python
#create model
model1 = Sequential()

#get number of columns in training data
n_cols = result.values.shape[1]

#add model layers
model1.add(Dense(100, activation='relu'))
model1.add(Dense(100, activation='relu'))
model1.add(Dense(50, activation='softmax'))

#compile model using mse as a measure of model performance
model1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

#set early stopping monitor so the model stops training when it won't improve anymore
early_stopping_monitor = EarlyStopping(patience=3)
#train model
model1.fit(result.values, df_output_grade_course_headers.values[:,0:50], epochs=10, validation_split=0.2)
```

```python
#create model
model2 = Sequential()

#get number of columns in training data
n_cols = result.values.shape[1]

#add model layers
model2.add(Dense(100, activation='relu'))
model2.add(Dense(200, activation='relu'))
model2.add(Dense(50, activation='softmax'))

#compile model using mse as a measure of model performance
model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

#set early stopping monitor so the model stops training when it won't improve anymore
early_stopping_monitor = EarlyStopping(patience=3)
#train model
model2.fit(result.values, df_output_grade_course_headers.values[:,50:100], epochs=10, validation_split=0.2, callbacks=[early_stopping_monitor])
```
